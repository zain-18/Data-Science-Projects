#### Max Depth

```python
from sklearn.metrics import accuracy_score,make_scorer
from sklearn.model_selection import cross_val_score
```

```python
acc_train=[]
acc_test=[]
acc_cv=[]
for i in range(3,9):
    dt= DecisionTreeClassifier(max_depth=i,random_state=100)
    dt.fit(X_train,y_train)
    
    pred_train=dt.predict(X_train)
    pred_test=dt.predict(X_test)
    cv=cross_val_score(dt,X_train,y_train,cv=5,scoring=make_scorer(accuracy_score))
    
    acc_train.append(accuracy_score(y_train,pred_train))
    acc_test.append(accuracy_score(y_test,pred_test))
    acc_cv.append(cv.mean())
print(acc_train)
print(acc_test)
print(acc_cv)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=range(3,9),y=acc_train,label='Training')
sns.lineplot(x=range(3,9),y=acc_test,label='Testing')
sns.lineplot(x=range(3,9),y=acc_cv,label='CV')
plt.show()
```

#### Min_SampleSplit

```python
acc_train=[]
acc_test=[]
acc_cv=[]
for i in range(3,9):
    dt= DecisionTreeClassifier(min_samples_split=i,random_state=100)
    dt.fit(X_train,y_train)
    
    pred_train=dt.predict(X_train)
    pred_test=dt.predict(X_test)
    cv=cross_val_score(dt,X_train,y_train,cv=5,scoring=make_scorer(accuracy_score))
    
    acc_train.append(accuracy_score(y_train,pred_train))
    acc_test.append(accuracy_score(y_test,pred_test))
    acc_cv.append(cv.mean())
print(acc_train)
print(acc_test)
print(acc_cv)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=range(3,9),y=acc_train,label='Training')
sns.lineplot(x=range(3,9),y=acc_test,label='Testing')
sns.lineplot(x=range(3,9),y=acc_cv,label='CV')
plt.show()
```
